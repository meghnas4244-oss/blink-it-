```python
import pandas as pd

# Blinkit Sales Dataset
data = {
    "Item Fat Content": ["Low Fat", "Regular", "Low Fat", "Regular", "Low Fat"],
    "Item Identifier": ["FDX01", "DRY02", "FDX03", "SNK04", "DRY05"],
    "Item Type": ["Dairy", "Soft Drinks", "Meat", "Snack Foods", "Hard Drinks"],
    "Outlet Establishment Year": [2005, 2010, 2005, 2018, 2010],
    "Outlet Identifier": ["OUT049", "OUT018", "OUT049", "OUT046", "OUT018"],
    "Outlet Location Type": ["Tier 1", "Tier 2", "Tier 1", "Tier 3", "Tier 2"],
    "Outlet Size": ["Medium", "Small", "Medium", "Large", "Small"],
    "Outlet Type": ["Supermarket Type1", "Supermarket Type2", "Supermarket Type1",
                     "Supermarket Type3", "Supermarket Type2"],
    "Item Visibility": [0.016, 0.045, 0.032, 0.089, 0.027],
    "Item Weight": [9.3, 12.5, 14.8, 7.6, 10.2],
    "Sales": [3735.14, 4434.42, 2097.27, 7322.55, 1589.26],
    "Rating": [4.2, 4.5, 4.0, 4.7, 4.1]
}

df = pd.DataFrame(data)

print(df)
```

```python
sales_outlet = df.groupby("Outlet Type")["Sales"].sum()

plt.figure()
plt.bar(sales_outlet.index, sales_outlet.values)
colors = ["brown", "pink", "maroon", "orange"]
plt.bar(sales_outlet.index, sales_outlet.values, color=colors)
plt.xlabel("Outlet Type")
plt.ylabel("Total Sales")
plt.title("Sales by Outlet Type")
plt.xticks(rotation=45)
plt.show()
```

```python
fat_sales = df.groupby("Item Fat Content")["Sales"].sum()

plt.figure()
plt.bar(fat_sales.index, fat_sales.values)
color=["#4CAF50", "#FF7043"],   #low fat = green, regular = orange
edgecolor="black"

plt.xlabel("Item Fat Content")
plt.ylabel("Total Sales")
plt.title("Sales by Item Fat Content")
plt.show()
```

```python
location_sales = df.groupby("Outlet Location Type")["Sales"].sum()

plt.figure()
plt.pie(location_sales.values, labels=location_sales.index, autopct="%1.1f%%")
plt.title("Sales Share by Outlet Location Type")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Item Visibility"], df["Sales"])
plt.xlabel("Item Visibility")
plt.ylabel("Sales")
plt.title("Item Visibility vs Sales")
plt.show()
```

```python
plt.figure()
plt.hist(df["Item Weight"])
plt.xlabel("Item Weight")
plt.ylabel("Frequency")
plt.title("Item Weight Distribution")
plt.show()
```

```python
rating_size = df.groupby("Outlet Size")["Rating"].mean()

plt.figure()
plt.bar(rating_size.index, rating_size.values)
plt.xlabel("Outlet Size")
plt.ylabel("Average Rating")
plt.title("Average Rating by Outlet Size")
plt.show()
```
